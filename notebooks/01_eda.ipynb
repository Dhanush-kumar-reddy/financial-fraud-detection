```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
sns.countplot(x='Class', data=df)
plt.title("Fraud vs Normal Transactions")
plt.show()
```

##### Insight
* Fraud is almost invisible
* Model will bias toward normal class

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Amount'], bins=50, kde=True)
plt.title("Transaction Amount Distribution")
plt.show()
```

##### Insight
* Highly right-skewed distribution
* Few very large transactions

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Class', y='Amount', data=df)
plt.title("Amount vs Fraud")
plt.show()
```

##### Insight
* Fraud transactions are not always high-value
* Pattern is non-trivial → requires ML

```python
plt.figure(figsize=(12,10))
corr = df.corr()

sns.heatmap(corr, cmap='coolwarm', center=0)
plt.title("Feature Correlation Matrix")
plt.show()
```

##### Insight

* Most features (V1–V28) are PCA → low correlation
* Hard to interpret directly

##### Observation : 
Since features are PCA-transformed, interpretability is limited, so I rely on model-based importance and SHAP later.

```python
corr_target = df.corr()['Class'].sort_values(ascending=False)

print(corr_target.head(10))
print(corr_target.tail(10))
```

##### Insight

* Some features strongly correlate (positive/negative)
* These are important signals

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Time'], bins=50)
plt.title("Transaction Time Distribution")
plt.show()
```

##### Insight

* Time is continuous
* Not directly meaningful → needs transformation

### EDA Insights

- Dataset is highly imbalanced (~0.17% fraud)

- PCA features reduce interpretability
